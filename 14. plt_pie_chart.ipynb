```python
#https://www.tutorialspoint.com/matplotlib/matplotlib_pie_chart.htm
from matplotlib import pyplot as plt  # Importing the pyplot module from the matplotlib library
import numpy as np  # Importing the numpy library and assigning it the alias np for numerical operations
```

```python
langs = ['C', 'C++', 'Java', 'Python', 'PHP']  # Defining programming languages as a list
students = [23, 17, 35, 29, 12]  # Defining the number of students for each language as a list
```

```python
fig = plt.figure()  # Creating a new figure object

ax = fig.add_axes([0, 0, 1, 1])  # Adding an axes object to the figure covering the entire figure

ax.axis('equal')  # Setting the aspect ratio of the axes to be equal, ensuring a circular pie chart

# Creating a pie chart with student data, labels, and displaying percentage values
ax.pie(students, labels=langs, autopct='%1.2f%%')

plt.show()  # Displaying the plot
```


ax.axis('equal'):

This line of code sets the aspect ratio of the axes to be equal. In the context of a pie chart, this is important to ensure that the pie chart appears as a circle rather than an ellipse. When the aspect ratio is equal, each unit on the x-axis is the same physical length as each unit on the y-axis, preserving the circular shape of the pie.

autopct='%1.2f%%':

This parameter is used when creating a pie chart to format the percentage labels displayed on each wedge of the pie.

    Explanation:
        autopct stands for "automatic percentage" and is used to format the percentage values.
        %1.2f%% is a formatting string:
            %1.2f specifies the format for floating-point numbers with at least one digit before the decimal point and two digits after the decimal point.
            %% is used to include a literal percentage sign.
        So, %1.2f%% means display the percentage values with one digit before the decimal point, two digits after the decimal point, and include a percentage sign.

    Example:
        If a wedge represents 25.1234% of the total, %1.2f%% would format it as "25.12%".
