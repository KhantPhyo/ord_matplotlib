```python
#https://www.tutorialspoint.com/matplotlib/matplotlib_twin_axes.htm
import matplotlib.pyplot as plt
import numpy as np
```

```python

# Generating an array x with values from 1 to 10
x = np.arange(1, 11)
```

```python
# Creating a new figure object
fig = plt.figure()

# Adding an axes object to the figure with the entire figure as its position and size
a1 = fig.add_axes([0, 0, 1, 1])

# Plotting the exponential function (exp(x)) on the left y-axis (a1)
a1.plot(x, np.exp(x))

# Setting the y-axis label for the left y-axis
a1.set_ylabel('exp')

# Creating a twinx axes, which shares the same x-axis with a1 but has a different y-axis
a2 = a1.twinx()

# Plotting the natural logarithm (log(x)) on the right y-axis (a2) with red line and circle markers
a2.plot(x, np.log(x), 'ro-')

# Setting the y-axis label for the right y-axis
a2.set_ylabel('log')

# Adding a legend to the figure with labels 'exp' and 'log' at the upper left corner
fig.legend(labels=('exp', 'log'), loc='upper left')

# Displaying the plot
plt.show()
```

```python
fig = plt.figure()
#left, bottom, width and height
a1 = fig.add_axes([0,0,1,1])
a1.plot(x,np.exp(x))
a1.set_ylabel('exp')
a1.set_ylim(0,10000)
a2 = a1.twinx()
a2.plot(x, np.log(x),'ro-')
a2.set_ylabel('log')
fig.legend(labels = ('exp','log'),loc='upper left')
plt.show()
```
