```python
#https://www.tutorialspoint.com/matplotlib/matplotlib_bar_plot.htm
import matplotlib.pyplot as plt  # Importing the matplotlib.pyplot module for plotting
import numpy as np
```

```python
# Defining the programming languages as a list
langs = ['C', 'C++', 'Java', 'Python', 'PHP']

# Defining the number of students for each language as a list
students = [23, 17, 35, 29, 12]
```

```python
fig = plt.figure()  # Creating a new figure object

# Adding an axes object to the figure with the entire figure as its position and size
ax = fig.add_axes([0, 0, 1, 1])

# Creating a bar plot on the axes with programming languages on the x-axis and number of students on the y-axis
ax.bar(langs, students)

# Displaying the plot
plt.show()
```

```python
data = [[30, 25, 50, 20],
        [40, 23, 51, 17],
        [35, 22, 45, 19]]
X = np.arange(4)  # Creating an array X with values from 0 to 3
```

data = [[30, 25, 50, 20], [40, 23, 51, 17], [35, 22, 45, 19]]: Defines a list of lists (data) where each inner list represents a set of data for a category.

X = np.arange(4): Creates an array X with values [0, 1, 2, 3]. This will be used as the x-axis positions for the bar charts.

```python
data[0]
```

```python
fig = plt.figure()  # Creating a new figure object

# Adding an axes object to the figure with the entire figure as its position and size
ax = fig.add_axes([0, 0, 1, 1])

# Plotting bar charts for each set of data
ax.bar(X + 0.00, data[0], color='b', width=0.25)  # Blue bars, offset by 0.00
ax.bar(X + 0.25, data[1], color='g', width=0.25)  # Green bars, offset by 0.25
ax.bar(X + 0.50, data[2], color='r', width=0.25)  # Red bars, offset by 0.50

# Displaying the plot
plt.show()
```

fig = plt.figure(): Creates a new figure object. A figure is like a blank canvas where you can place your plots.

ax = fig.add_axes([0, 0, 1, 1]): Adds an axes object (ax) to the figure, covering the entire figure. The parameters [0, 0, 1, 1] represent [left, bottom, width, height], and in this case, the axes cover the entire figure.

ax.bar(X + 0.00, data[0], color='b', width=0.25): Plots the first set of bars for the first category. The bars are blue ('b'), offset by 0.00 on the x-axis, and have a width of 0.25.

ax.bar(X + 0.25, data[1], color='g', width=0.25): Plots the second set of bars for the second category. The bars are green ('g'), offset by 0.25 on the x-axis, and have a width of 0.25.

ax.bar(X + 0.50, data[2], color='r', width=0.25): Plots the third set of bars for the third category. The bars are red ('r'), offset by 0.50 on the x-axis, and have a width of 0.25.

plt.show(): Displays the plot. In this case, it shows a grouped bar chart with three sets of bars representing different categories. Each category has four bars, and the bars are grouped together for each position on the x-axis.

```python
fig = plt.figure()  # Creating a new figure object

# Adding an axes object to the figure with the entire figure as its position and size
ax = fig.add_axes([0, 0, 1, 1])

# Plotting bar charts for each set of data
ax.bar(X + 0.00, data[0], color='b', width=0.25)  # Blue bars, offset by 0.00
ax.bar(X + 0.15, data[1], color='g', width=0.25)  # Green bars, offset by 0.15
ax.bar(X + 0.30, data[2], color='r', width=0.25)  # Red bars, offset by 0.30

# Displaying the plot
plt.show()
```

```python
N = 5  # Number of groups
menMeans = (20, 35, 30, 35, 27)  # Mean scores for men in each group
womenMeans = (25, 32, 34, 20, 25)  # Mean scores for women in each group
ind = np.arange(N)  # Generating an array of x locations for the groups
width = 0.35  # Width of the bars
```

```python
fig = plt.figure()  # Creating a new figure object

# Adding an axes object to the figure with the entire figure as its position and size
ax = fig.add_axes([0, 0, 1, 1])

# Plotting bar chart for men's scores
ax.bar(ind, menMeans, width, color='r')

# Plotting bar chart for women's scores, stacked on top of men's scores
ax.bar(ind, womenMeans, width, bottom=menMeans, color='b')

ax.set_ylabel('Scores')  # Setting the y-axis label
ax.set_title('Scores by group and gender')  # Setting the title of the plot

# Setting the x-axis ticks and labels
ax.set_xticks(ind)
ax.set_xticklabels(['G1', 'G2', 'G3', 'G4', 'G5'])

# Setting the y-axis ticks
ax.set_yticks(np.arange(0, 81, 10))

# Adding a legend to the plot with labels 'Men' and 'Women'
ax.legend(labels=['Men', 'Women'])

# Displaying the plot
plt.show()
```
