```python
#https://www.tutorialspoint.com/matplotlib/matplotlib_histogram.htm
import matplotlib.pyplot as plt
import numpy as np
```

```python
a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27])
```

ax.hist(a, bins=[0, 25, 50, 75, 100])

In this example, you explicitly define the bin edges. The bins are defined as intervals: [0, 25), [25, 50), [50, 75), [75, 100). The last bin is right-closed, which means it includes values up to 100.

    Advantages:
        You have explicit control over the bin edges.
        Useful when you have specific intervals or thresholds you want to highlight.

    Disadvantages:
        You need to manually specify bin edges, which may be cumbersome if the data range is large or if you are not sure about the appropriate bin edges.

```python
fig,ax = plt.subplots(1,1)
ax.hist(a, bins = [0,25,50,75,100])
ax.set_title("histogram of result")
ax.set_xticks([0,25,50,75,100])
ax.set_xlabel('marks')
ax.set_ylabel('no. of students')
plt.show()
```

Here's how bins work in a histogram:

    Determine the Range of Data:
        First, you need to determine the range of your data, i.e., the minimum and maximum values.

    Divide the Range into Bins:
        The range is then divided into intervals or bins. Each bin represents a specific range of values.

    Count Data Points in Each Bin:
        For each data point, determine the bin it falls into. Increment the count for that bin.

    Plotting the Histogram:
        The histogram is then plotted with bars representing the bins and their corresponding frequencies.

```python
import matplotlib.pyplot as plt  # Importing the matplotlib.pyplot module for plotting
import numpy as np  # Importing the numpy library and assigning it the alias np for numerical operations

# Generate random data for the histogram (1000 random samples from a standard normal distribution)
data = np.random.randn(1000)

# Plotting a basic histogram with 30 bins, skyblue color, and black edge color
plt.hist(data, bins=30, color='skyblue', edgecolor='black')

# Adding labels to the x and y axes
plt.xlabel('Values')
plt.ylabel('Frequency')

# Adding a title to the histogram
plt.title('Basic Histogram')

# Display the plot
plt.show()

```

plt.hist(data, bins=30, color='skyblue', edgecolor='black')

In this example, you specify the number of bins (bins=30) rather than the explicit bin edges. Matplotlib will automatically divide the range of the data into 30 bins.

    Advantages:
        Convenient when you want a quick visualization without specifying bin edges.
        Matplotlib handles the binning process for you.

    Disadvantages:
        You may not have explicit control over the exact bin edges, which might be important in some cases.
