```python
#https://www.tutorialspoint.com/matplotlib/matplotlib_setting_limits.htm
import matplotlib.pyplot as plt  # Importing the matplotlib.pyplot module for plotting
import numpy as np  # Importing the numpy library and assigning it the alias np for numerical operations12
```

```python
x = np.arange(1, 10)  # Creating an array x with values from 1 to 9
```

```python
fig = plt.figure()  # Creating a new figure object

a1 = fig.add_axes([0, 0, 1, 1])  # Adding an axes object to the figure with specified position and size
# The parameters [0, 0, 1, 1] represent [left, bottom, width, height], where (0, 0) is the bottom-left corner,
# and (1, 1) is the top-right corner, so the axes cover the entire figure.
a1.plot(x, np.exp(x))  # Plotting x versus e^x on the specified axes (a1)
a1.set_title('exp')  # Setting the title for the plot

plt.show()  # Displaying the plot
```

```python
fig = plt.figure()  # Creating a new figure object

# Adding an axes object to the figure with specified position and size
# The parameters [0, 0, 1, 1] represent [left, bottom, width, height]
a1 = fig.add_axes([0, 0, 1, 1])

# Plotting x versus e^x on the specified axes (a1) with a red line ('r')
a1.plot(x, np.exp(x), 'r')

a1.set_title('exp')  # Setting the title for the plot

# Setting the y-axis limits for the plot to be between 0 and 10000
a1.set_ylim(0, 10000)

# Setting the x-axis limits for the plot to be between 0 and 10
a1.set_xlim(0, 10)

plt.show()  # Displaying the plot
```
