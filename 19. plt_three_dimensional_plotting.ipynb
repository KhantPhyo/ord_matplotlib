```python
#https://www.tutorialspoint.com/matplotlib/matplotlib_three_dimensional_plotting.htm
from mpl_toolkits import mplot3d  # Importing the 3D plotting toolkit from matplotlib
import numpy as np  # Importing the numpy library and assigning it the alias np for numerical operations
import matplotlib.pyplot as plt  # Importing the pyplot module from the matplotlib library
```

```python
#https://en.wikipedia.org/wiki/Circle
z = np.linspace(0, 1, 100)  # Generating an array of 100 evenly spaced values from 0 to 1
x = z * np.sin(20 * z)  # Calculating x values based on the sine function and the z values
y = z * np.cos(20 * z)  # Calculating y values based on the cosine function and the z values
x.shape, y.shape  # Checking the shape of the x and y arrays
```

```python
fig = plt.figure()  # Creating a new figure object

ax = plt.axes(projection='3d')  # Creating a 3D axes instance

ax.plot3D(x, y, z, 'gray')  # Creating a 3D line plot with gray color
ax.set_title('3D line plot')  # Setting the title of the plot

plt.show()  # Displaying the plot
```

```python
c = x + y
c.shape
```

```python
ax = plt.axes(projection='3d')
ax.scatter(x, y, z, c=c)
ax.set_title('3d Scatter plot')
plt.show()
```
