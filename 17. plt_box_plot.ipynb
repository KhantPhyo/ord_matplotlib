```python
#https://www.tutorialspoint.com/matplotlib/matplotlib_box_plot.htm
import matplotlib.pyplot as plt  # Importing the pyplot module from the matplotlib library
import numpy as np  # Importing the numpy library and assigning it the alias np for numerical operations
```

```python
np.random.seed(10)  # Setting a seed for reproducibility of random numbers

# Generating random data for four collections using normal distribution
collectn_1 = np.random.normal(100, 10, 200)  # Mean=100, Standard Deviation=10, 200 data points
collectn_2 = np.random.normal(80, 30, 200)   # Mean=80, Standard Deviation=30, 200 data points
collectn_3 = np.random.normal(90, 20, 200)   # Mean=90, Standard Deviation=20, 200 data points
collectn_4 = np.random.normal(70, 25, 200)   # Mean=70, Standard Deviation=25, 200 data points
```

```python
# Creating a list containing the four collections of random data
data_to_plot = [collectn_1, collectn_2, collectn_3, collectn_4]
```

```python
#https://en.wikipedia.org/wiki/Box_plot
#https://www.youtube.com/watch?v=6YDHBFVIvIs
#https://www.thoughtco.com/calculate-a-sample-standard-deviation-3126345

fig = plt.figure()  # Creating a new figure object

# Creating an axes instance covering the entire figure
ax = fig.add_axes([0, 0, 1, 1])

# Creating a boxplot using the boxplot function with the four collections of data
bp = ax.boxplot(data_to_plot)

plt.show()  # Displaying the plot
```
