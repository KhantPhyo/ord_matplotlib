```python
#https://www.tutorialspoint.com/matplotlib/matplotlib_multiplots.htm
import matplotlib.pyplot as plt  # Importing the matplotlib.pyplot module for plotting
import numpy as np  # Importing the numpy library and assigning it the alias np for numerical operations
```

```python
plt.plot([1,2,3])  # Plotting a simple line with values [1, 2, 3] that mean, x is index - 0, 1, 2 and y is value - 1, 2 ,3
```

```python
# Plotting two subplots in a grid with 2 rows and 1 column
plt.subplot(211)  # Creating the first subplot in a 2x1 grid
plt.plot(np.arange(12))  # Plotting a line in the first subplot

plt.subplot(212, facecolor='y')  # Creating the second subplot in a 2x1 grid with a yellow background
plt.plot(np.arange(12))  # Plotting a line in the second subplot
```

1x1 Grid
(1, 1, 1): Single subplot in a 1x1 grid.

2x1 Grid
(2, 1, 1): Subplot in the top position of a 2x1 grid.
(2, 1, 2): Subplot in the bottom position of a 2x1 grid.

(2, 2, 4): Subplot in the bottom-right position of a 2x2 grid.

(3, 1, 2): Subplot in the middle position of a 3x1 grid.
(3, 1, 3): Subplot in the bottom position of a 3x1 grid.

3x2 Grid

```python
fig = plt.figure()  # Creating a new figure

# Creating a subplot with a single plot in the figure
ax1 = fig.add_subplot(111) # create ax1 object instance for OOP Feature (if we want, we can use directly - plt)
ax1.plot([1,2,3])  # Plotting a line in the single subplot
```

```python
fig = plt.figure()  # Creating a new figure

# Creating a subplot with a 2x2 grid and selecting the first subplot
ax2 = fig.add_subplot(221, facecolor='y')
ax2.plot([1,2,3])  # Plotting a line in the first subplot of the 2x2 grid
```

```python
fig = plt.figure()  # Creating a new figure

# Creating a subplot with a single plot in the figure
ax1 = fig.add_subplot(111)
ax1.plot([1,2,3])  # Plotting a line in the single subplot

# Creating a subplot with a 2x2 grid and selecting the second subplot
ax2 = fig.add_subplot(221, facecolor='y')
ax2.plot([1,2,3])  # Plotting a line in the second subplot of the 2x2 grid
```

```python
x = np.arange(0, np.pi*2, 0.05)  # Generating an array of values from 0 to 2 with a step size of 0.05
y1 = np.sin(x)  # Calculating the sine of each value in the array x
y2 = np.cos(x)  # Calculating the cosine of each value in the array x
```

```python
fig = plt.figure()  # Creating a new figure

# Adding the main subplot to the figure with a set position and size (left, bottom, width, height)
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])

# Adding an inset subplot to the figure with a different position and size
axes2 = fig.add_axes([0.55, 0.55, 0.3, 0.3])

# Plotting the sine function in the main subplot
axes1.plot(x, y1, 'b')

# Plotting the cosine function in the inset subplot
axes2.plot(x, y2, 'r')

# Setting titles for the main and inset subplots
axes1.set_title('sine')
axes2.set_title("cosine")

plt.show()  # Displaying the figure with subplots
```

fig.add_axes(): This method is used to add an axes (subplot) to a figure. It takes a list of four values [left, bottom, width, height] that define the position and size of the subplot.

left: The position of the left edge of the subplot in normalized figure coordinates. Here, 0.1 means 10% from the left.

bottom: The position of the bottom edge of the subplot in normalized figure coordinates. Here, 0.1 means 10% from the bottom.

width: The width of the subplot in normalized figure coordinates. Here, 0.8 means 80% of the figure width.

height: The height of the subplot in normalized figure coordinates. Here, 0.8 means 80% of the figure height.

So, axes1 is a subplot that covers 80% of the width and 80% of the height of the figure, starting from 10% from the left and 10% from the bottom.

____________________________________________________________________

Similar to the explanation above, this line adds another subplot (axes2) to the figure.

The [0.55, 0.55, 0.3, 0.3] specifies the position and size of the inset subplot.

0.55 for left and 0.55 for bottom mean that the inset subplot starts at 55% from the left and 55% from the bottom.

0.3 for width and 0.3 for height mean that the inset subplot has a width and height of 30% of the figure.

So, axes2 is an inset subplot that covers 30% of the width and 30% of the height of the figure, starting from 55% from the left and 55% from the bottom.

In summary, these lines are creating two subplots (axes1 and axes2) in a single figure, where axes1 is the main subplot, and axes2 is an inset subplot.
