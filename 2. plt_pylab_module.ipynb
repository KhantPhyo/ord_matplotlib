```python
#https://www.tutorialspoint.com/matplotlib/matplotlib_pylab_module.htm
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-3, 3, 30)
print(x)
y = x**2
print(y)
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y, 'r.')
plt.show()
```

```python
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x), 'r-')
plt.plot(x, -np.sin(x), 'g--')
plt.show()
```
