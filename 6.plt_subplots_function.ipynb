```python
#https://www.tutorialspoint.com/matplotlib/matplotlib_subplots_function.htm
import matplotlib.pyplot as plt  # Importing the matplotlib.pyplot module for plotting
import numpy as np  # Importing the numpy library and assigning it the alias np for numerical operations
```

```python
x = np.arange(1, 5)  # Creating an array x with values from 1 to 4 (exclusive)
print("X  is : ",x)
```

plt.subplots(nrows, ncols): This function creates a new figure and a grid of subplots.

nrows: This parameter specifies the number of rows in the grid. In the example above, nrows=2, so there are 2 rows of subplots.

ncols: This parameter specifies the number of columns in the grid. In the example above, ncols=2, so there are 2 columns of subplots.

The function returns two values: fig and axs.

fig: This is the Figure object representing the entire figure, or the top-level container for all the subplots.

axs: This is an array or a 2D array of Axes objects. Each element in the array corresponds to a subplot. In the example above, axs would be a 2D array with dimensions 2x2 because there are 2 rows and 2 columns.



```python
# Creating a 2x2 subplot grid and getting the figure and axes objects
fig, ax = plt.subplots(2, 2)  # 'ax' is a 2D array of axes objects (2 rows, 2 columns)

# Plotting x versus x squared in the first subplot (top-left)
ax[0][0].plot(x, x*x)  # First row, first column
ax[0][0].set_title('square')  # Setting the title for the subplot

# Plotting x versus square root of x in the second subplot (top-right)
ax[0][1].plot(x, np.sqrt(x))  # First row, second column
ax[0][1].set_title('square root')  # Setting the title for the subplot

# Plotting x versus e^x (exponential) in the third subplot (bottom-left)
ax[1][0].plot(x, np.exp(x))  # Second row, first column
ax[1][0].set_title('exp')  # Setting the title for the subplot

# Plotting x versus log base 10 of x in the fourth subplot (bottom-right)
ax[1][1].plot(x, np.log10(x))  # Second row, second column
ax[1][1].set_title('log')  # Setting the title for the subplot

# Displaying the entire set of subplots
plt.show()
```
